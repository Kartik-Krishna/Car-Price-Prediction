```python
import selenium
import pandas as pd
from selenium import webdriver
import time
import numpy as np
import matplotlib.pyplot as plt               
import seaborn as sns
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
url = 'https://www.olx.in'
driver.get(url)

time.sleep(3)

cars_page = driver.find_element_by_xpath('//div[@class="_3AGJR _18NX_"][1]//a')
cars_page.click()
time.sleep(2)

```

```python
loc_href = []
href = []
loc_tags = driver.find_elements_by_xpath('//a[@class="_396bV _2zs4P _1nXYG"]')

for i in loc_tags:
    loc_href.append(i.get_attribute('href'))

for i in loc_href:
    driver.get(i)
    time.sleep(2)
    
    load_more = driver.find_element_by_xpath('//button[@class="rui-39-wj rui-3evoE rui-1JPTg"]')
    load_more.click()
    time.sleep(2)
    driver.find_element_by_tag_name('body').send_keys(Keys.CONTROL + Keys.END)
    
    load_more = driver.find_element_by_xpath('//button[@class="rui-39-wj rui-3evoE rui-1JPTg"]')
    load_more.click()
    time.sleep(2)
    
    driver.find_element_by_tag_name('body').send_keys(Keys.CONTROL + Keys.HOME)
    get_href = driver.find_elements_by_xpath('//li[@class="EIR5N"]//a')
    
    for i in get_href:
        href.append(i.get_attribute('href'))
        
print (href)
len (href)
```

```python
brand = []
fuel = []
km = []
model = []
year = []
variant = []
transmission = []
no_own = []
price = []
loc = []

for i in href:
    driver.get(i)
    time.sleep(3)
    
    try:
        price_tag = driver.find_element_by_xpath('//span[@class="_2xKfz"]')
        price.append(price_tag.text)
    
    except:
        price.append('-')
        
    try:
        loc_tag = driver.find_element_by_xpath('//span[@class="_2FRXm"]')
        loc.append(loc_tag.text)
        
    except:
        loc.append('-')
    
    try:
        brand_tag = driver.find_element_by_xpath('//span[@data-aut-id="value_make"]')
        brand.append(brand_tag.text)
        
    except:
        brand.append('-')
        
    try:
        fuel_tag = driver.find_element_by_xpath('//span[@data-aut-id="value_petrol"]')
        fuel.append(fuel_tag.text)
        
    except:
        fuel.append('-')
        
    try:
        km_tag = driver.find_element_by_xpath('//span[@data-aut-id="value_mileage"]')
        km.append(km_tag.text)
        
    except:
        km.append('-')
        
    try:
        model_tag = driver.find_element_by_xpath('//span[@data-aut-id="value_model"]')
        model.append(model_tag.text)
        
    except:
        model.append('-')
        
    try:
        year_tag = driver.find_element_by_xpath('//span[@data-aut-id="value_year"]')
        year.append(year_tag.text)
        
    except:
        year.append('-')
        
    try:
        var_tag = driver.find_element_by_xpath('//span[@data-aut-id="value_variant"]')
        variant.append(var_tag.text)
        
    except:
        variant.append('-')
        
    try:
        tran_tag = driver.find_element_by_xpath('//span[@data-aut-id="value_transmission"]')
        transmission.append(tran_tag.text)
        
    except:
        transmission.append('-')
        
    try:
        own_tag = driver.find_element_by_xpath('//span[@data-aut-id="value_first_owner"]')
        no_own.append(own_tag.text)
        
    except:
        no_own.append('-')
        

print (brand)
print (fuel)
print (km)
print (model)
print (year)
print (variant)
print (transmission)
print (no_own)
print (price)
print (loc)
```

```python
olx = pd.DataFrame({})
olx['Brand'] = brand
olx['Model'] = model
olx['Variant'] = variant
olx['Year'] = year
olx['Fuel'] = fuel
olx['Transmission'] = transmission
olx['Km Driven'] = km
olx['No. of Owners'] = no_own
olx['Location'] = loc
olx['Price'] = price
```
